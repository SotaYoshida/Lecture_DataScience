# Pythonの環境構築

ローカル環境にPythonを導入する方法をOSごとに紹介する。  
難易度としてはLinux < Mac << (壁)<<  Windowsといった感じ(私見).

## Windowsの場合

いくつかの方針が考えられる。

1. Pythonのインストーラを用いてインストールする  
  * メリット: Windows環境を汚さない, インストールが楽  
  * デメリット: Linuxシェル環境が使えない。吉田が発狂する。  

2. Linux環境を構築する
* 2-1.   Windows Subsystem for Linux(WSL)  
  * メリット: Windows環境を汚さない, Linuxシェル環境が使える  
  * デメリット: WSL特有の情報を調べる必要がある

* 2-2.  デュアルブート   
  * メリット: Windowsとは別に好きなLinuxディストリビューションを共存させられる  
  (Ubuntuはユーザー数がそれなりに多くUIがカッコいい)  
  * デメリット: ディスク領域の分割などが必要

吉田はWindows環境を長時間使うと発作(Windowsディス)が出るので、  
1.の方法を選択する方は注意(?)してください.

#### 手順4. 
Ubuntuターミナルに以下を入力し実行
```

sudo apt upgrade
```
sudoは管理者権限で実行する、という意味で、  
aptはLinuxのパッケージ管理システムです。  
※PCの時刻設定がきちんと行われていないと、aptのupdateやupgradeに失敗します。


ライブラリ・モジュールをインストールしたければ
```

pip3 install pandas
pip3 install Selenium
```

## Linuxの場合

Linuxを使っている人が、pythonのインストールができないとは思いにくいが...  

ターミナルから
```
sudo apt update -y
sudo apt upgrade -y
sudo apt dist-upgrade -y
sudo apt autoremove -y
```
などを実行しパッケージマネージャのアップデート等を行った後

```

```

などを実行すればよい。

pipがない場合、
```
sudo apt install python3-pip
```
を実行する。

## Linux/Unix ターミナルの操作

基本的な事項をまとめる

* Ctrl+C: 入力内容の消去・実行中の作業の中断
* Ctrl+D: ターミナル(タブ)を閉じる (Pythonの対話モードから出るときにも使う)
* Tab: 入力を補完してくれる

基本的なコマンド
* cd チェンジディレクトリ, "現在地"を変更する
```
cd ./Desktop
cd ../
```

* ls リストコマンド, ファイルやディレクトリなどを表示する.
```
ls 
ls ../*
ls Desktop/*
``` 

* mkdir コマンド (ディレクトリを作る)
``` 
mkdir Desktop/AdDS2020
```

* rm ファイルやディレクトリを削除する
```
rm ./Desktop/report1.txt
rm -r ./Desktop/AdDS2020
```
通常、rmコマンドで削除したものはゴミ箱に入れられずに削除されるので注意  
(ゴミ箱を経由するよう設定を変更するのが推奨)

* grep 検索
``` 
grep "Hello World" Desktop/AdDS2020/*py
```
