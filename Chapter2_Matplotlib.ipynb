### 棒グラフ その１

Aさんのセンター試験の得点をリスト(括弧[ ]内に、数字カンマで区切る)に格納する。  
順番(各科目の名前(ラベル))は、国語,英語,数IA,数IIB,化学,物理,世界史とする。  

```
data = [152, 170, 82,85,79,92,88]
label = ["国語","英語", "数IA", "数IIB", "化学", "物理", "世界史"]
total = sum(data)
print("合計得点は", total, "点で、得点率は"+str("%6.1f" % (100* total/900))+"%です")
```

```
plt.figure(figsize=(7,2)) ## matplotlib のfigureという関数を読み込んで、キャンバスを用意する figsize=(横,縦)で大きさが指定できる。
plt.bar(label,data,align='center',width=0.5,color="red") ## bar(棒)グラフを描く。1つめの引数をx軸(数値以外のときは整数値にしてくれる),2つめの引数をy軸にplotする。
plt.xlabel("各科目"); plt.ylabel("Aさんの得点") ## x軸とy軸のデータの種類(ラベル,label)を指定する
plt.show() ## それまでに指定した条件で絵を描く。
plt.close() ###　キャンバスを閉じる。　一つのプログラムで複数絵を描くときはこれを書く。
```

#### 2次元データのプロット

xとy、２つの量があって、どちらにも興味がある、あるいは両者の間の相関に興味がある場合、
散布図を描くと、見通しやすくなることが多いです。

以下では、2017年の宇都宮の平均気温とアイスクリームの消費量の相関を見てみましょう。
気温・アイスクリームの消費量ともに、1月から12月に順番にリストに入れていくことにします。



```
x= [3.1, 4.3, 6.6, 13.2, 19.1, 20.9, 26.4, 25.1, 21.9, 15.7, 9.6, 3.8]
y= [568, 572, 804, 833, 930, 965, 1213, 1120, 835, 540, 451, 502]

plt.figure(figsize=(4,4)) 
plt.title("宇都宮市") ## 図にはタイトルをつけることができます
plt.xlabel("平均気温 (℃)")
plt.ylabel("世帯あたりのアイスクリム・シャーベットの消費金額 (円)")
plt.scatter(x,y)
plt.show()
plt.close() 
```

せっかく相関を勉強したので、相関係数などの情報を含んだもう少しかっこいい図を作ってみましょう。

```
import numpy as np ## numpy というライブラリをnpと言う名前でインポート
x= np.array(x); y= np.array(y)## 注1　
r = np.corrcoef(x,y) ## x,yの配列間の相関行列を計算します 
corrcoef=r[0,1] ## 行列の非対角要素が相関係数です。 r[0,1] # 0行目1列の値

def seasoncolor(month):
    if month <= 2 or month ==12:
        return "blue"
    elif 3 <= month <=5:
        return "green"
    elif 6 <= month <=8:
        return "red"
    elif 9<= month <=11:
        return "orange"
    return tcol

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1,1,1)  ## 注2
ax.set_facecolor("#D3DEF1")
ax.set_title("宇都宮市")
ax.set_xlabel("平均気温 (℃)")
ax.set_ylabel("世帯あたりのアイスクリム・シャーベットの消費金額 (円)")
ax.grid(True,axis="both",color="w", linestyle="dotted", linewidth=0.8)
for i in range(len(x)):
    tcol=seasoncolor(i+1)
    ax.scatter(x[i],y[i],marker="o",s=10,color=tcol,zorder=20000,alpha=0.7)
    ax.text(x[i],y[i],str(i+1)+"月",color="k",fontsize=8)
ax.text(0.1,0.9, "r="+str("%5.2f" % corrcoef), transform=ax.transAxes,fontsize=12)
plt.show()
plt.close()
```

注1: x,yのデータを np.array()という型に変換しました.
今の場合はnp.array()にする必要性はないのですが、こうしておくと数学的な処理(ベクトル積,行列操作など)が高速になるため、[数値を格納したリストは常にnp.arrayに変換して使う]と慣れるまでは考えてもよいかもしれません。  
たとえばx軸の気温を絶対温度(ケルビン)に直したい時、リストのままでは、各要素に-273.15を加える必要がありますが、気温のリストをnp.arrayにしておけば


```
x + -273.15
```

とするだけで、全ての要素に-273.15を加えたnp.arrayを返します。

注2: axという見慣れないものが導入されました。axは、plt.figure()で用意したキャンバスの上に、
作業領域を指定するための関数で、たとえばキャンバスを四分割して、似たようなグラフを4つ同時に描いたりするのに便利です。

ただし、axなどを使いこなすのは少々テクニカルな点も多いので、よくわからない部分はとりあえず飛ばし読みで構いません。慣れてくると、その便利さに気がつくはずです。
