### 変数の定義と簡単な演算

変数に値(実数値,整数)を代入してみます。  

プログラムでは通常等号(=)は代入に使用されます。   
値の代入は、等号(＝)の左に変数(値を入れたいものの名前)、右に値を書くことで行います。
[変数] = [値]

````
a = 2.0 
b = 5
````

これでa,bに値が代入されました。

### 簡単な演算
次にこれらを使って、四則演算をしてみましょう

足し算

````
a+b 
````

引き算

````
a-b
````

掛け算 掛け算記号は*(アスタリスク)です

````
a*b
````

演算した結果を別の変数cに代入して保存しておくこともできます。

````
c=a*b # このように、新たに値を代入した場合はc(=a*b)の値は表示されません。  
````

cの値が後で知りたいと思ったときは、以下のようにprint関数というものを使います。

````
print(c)
````

以下ではprint関数を逐次使って、値や演算の結果を確認することにしましょう。  
またコード部分に#(半角シャープ記号)をつけると、コンピュータは#以降に書かれたことを無視しますので、コメントを書くことができます。

なれないうちはどういった計算をするのか、その行をいつ編集したのかなどコメントをいちいち書いておくと便利かもしれません。

````
print(b//a) # 切り捨て除算
print(b%a) # bをaで割った余り
print(a**b) # aのb乗
````

```a = a + 2```  
は、aに2を加えたものでaを再定義することを表します。
また、  
```a += 2 ```  
と書いても同じ結果が得られます。
+= を用いた書き方のほうがシンプルなので、今後多用します。

````
a = 2.0 
a += 4.0; print("二行目", a)
a += -3.0; print("三行目", a)
````

**約束 変数名に全角文字(ひらがな・カタカナ・漢字)はできるだけ使わない!!**

````
あ=1.0;print(あ) ## 非推奨
````

上記のように、変数にひらがな等を使用することは可能です。  
Google Colaboratoryでは何も特別なことをしなくてもひらがなが使えますし,  
最近の多くのプログラミング言語は全角文字をサポートしています。  
ただOS間の互換性などを考慮すると、一般にはひらがなや漢字など全角文字を変数に指定することは避けたほうが無難です。  
ユニコードでのプログラミングに馴染みがない偉い先生に全角スペースや全角文字が入ったコードを渡すと...怒られが発生しない保証はありません。  

以降では、変数名に全角文字は使いません。  
変数名は基本的に半角英数字, アンダースコア"_", のみで定義しましょう。  
```+,-,%,/,=```などの記号も使うことができません。  
また変数の頭に数字を使うこともできません。


### Pythonでの文字列操作

Pythonでは、数値はもちろん文字列も扱うことができます。  
比較的簡単に文字列の操作ができることも、Pythonが愛される理由の一つでしょう。


````
text = "私の名前はXXXです" ## XXXのところを自分の名前にしてみましょう。
print(text)
````

文字列と文字列同士では、足し算や掛け算が適用できます。
たとえば


````
text1 = "ティッシュ配りの"
text2 = "バイトを監視する"
a = 2
### 足し算
print(text1 + text2) 
````

とすれば、text1とtext2にそれぞれ代入した２つの文字列を連結した一つの文字列として表示することができます。
また、

````
print(str(a)+"人の"+text1+ text2 * 5 +"正社員")
````

とすれば、text2部分を５回繰り返した長い文字列を作ることができます。  
ここで、整数値を代入したaを文字列として他の文字列と結合するときは、str()関数(string関数/ストリング関数)を使って、文字列に変換してやらなくてはいけません。詳しくは後で"型"というものを勉強します。

#### 予約語

Pythonには、あらかじめ役割が与えられている言葉=予約語があり、以下は変数名として使う事はできません。

False, None, True, and, as, assert, break, class, continue, def, del, elif, else, except, finally, for, from, global, if, import, in, is, lambda, nonlocal,not, or, pass, raise, return, try, while, with, yield



これらの役割については登場したその都度説明します。  

実際にコード用のセルにこれらの語句を入力すると、黒ではなく別の色で表示されます。  
こういう場合は、変数に別の名前をつけるようにしましょう。
