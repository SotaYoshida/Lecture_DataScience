#### ファイルの参照やその他の使い方

Google Colab.では先頭に!をつけることでターミナルで使用するコマンドを実行することもできます。  
つまり、リストの参照をしたいときは!lsとすればよく、ディレクトリを新たに作成したいときは!mkdirといった具合です.

````
!ls ./gdrive/My\ Drive/pdf/*.pdf ## pdfディレクトリ内の.pdfで終わるファイルの一覧を表示するコマンド
!mkdir ./gdrive/My\ Drive/pdf #マイドライブ直下にpdfというディレクトリを作成するコマンド マイドライブ内にpdfというディレクトリがないときはこれを叩く
````

````
from pdf2image import convert_from_path
from PIL import Image
import cv2
import numpy as np

inpf = "input fileの名前(住所)を指定"
oupf = inpf.replace(".pdf","_resize.pdf")
dpi = 300 ## 解像度に相当します。必要に応じて変更してください
image = convert_from_path(inpf)
image = np.array(image)
image = Image.fromarray(image)
image.save(oupf,resolution=dpi)
````
