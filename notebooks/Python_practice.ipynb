### テンプレ課題1: オープンデータの収集と可視化

政府統計などのオープンデータをなるべく多く取得して、可視化したり考察をしてみよう。

 1. まずは、ほしいデータの種類と収集方法を考えよう。政府統計ならe-Stat,他にも各地方自治体などがデータを提供していることも多い。
 2. データの種類と収集先の目処が経ったら、まずは利用規約を読んでスクレイピング等を禁止していないかを調べる。
 3. 問題がないことを確認したら、データの収集にあたる。APIなどのデータ取得用の窓口があれば必ずそちらを利用する。
 4. 次に得られたデータを時系列や他のデータとの相関、年代ごとの棒グラフなど様々な方法で可視化する。
 5. 可視化されることで得られる理解や予想、その背景について考察を行う。

### テンプレ課題2: 機械学習を用いた画像認識/画像生成

1. 画像認識や画像生成を行う機械学習モデルについて概観し、授業を受講している他の学生に説明できるよう解説する。
    その際、情報の参照元は論文が望ましいが、難しければ日本語等の解説を参照して構わない。(注:もちろんレポート内では適切に引用・参照すること)

2. 次に、それらの機械学習モデルの中から、PythonがインターフェースとなっていてGoogleColab上で容易に実行ができそうなものを選ぶ。例えば...
    - [YOLOv5](https://github.com/ultralytics/yolov5): CNNを用いた画像認識のパッケージ
    - [StableDiffusion](https://github.com/CompVis/stable-diffusion): 潜在拡散モデルを用いた画像生成モデル
3. これらを実際に動かして画像認識や画像生成を実行してみる。
4. 関連するサービスや企業での導入事例などを調べると同時に、機械学習モデルによる画像認識/画像生成を用いる上で起こり得るトラブルや問題点について記述する。

### テンプレ課題3: 簡単なゲーム開発

数あてゲームやブラックジャックやババ抜きといったトランプを用いたゲーム、オセロなどを作る。
(Pythonにはゲーム作成のためのライブラリなどもあるが、これらは自由課題にあたる)

1. そのために、まずはターンや盤面の状態を表現するための方法を考えよう。例えば...
    - 山札/場、人数分のリストを用意しておいて手札や盤面などを表現する
    - ターンが進行するたびに絵をplotする仕組みにして、視覚的に表現する

    など、考え方は色々ある。参考までに、コードの中に適宜入力を受け取るプログラムを記述しておけば入力に応じて
2. ゲーム開始の状況を表現するための方法を考える(山札をランダムに配る, 駒の初期配置など)
