2010年の10週目など、数個以外はかなりいい感じにデータが取得できたのではないでしょうか？

```
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

#years = range(2010,2020)
years = range(2010,2011)
sw = 46; ew = 22
weeks = [j for j in range(sw,53)] + [i for i in range(1,ew)]

sumV=[]

for s_year in years:
    for week in weeks:
        if week < 10:
            tweek = "0"+str(week)
        else:
            tweek = str(week)
        if s_year < 2007:
            tw= str(tweek)
        else:
            tw=""
        if week <=ew :
            c_year = s_year+1
        else:
            c_year = s_year
        url = 'https://nesid4g.mhlw.go.jp/Hasseidoko/Levelmap/flu/'+str(s_year)+'_'+str(s_year+1)+'/'+str(c_year)+'_'+str(tweek)+'/jmap'+str(tw)+'.html'
        ctext = str(s_year)+"年"+tweek+"週目: "
        try :
            response = requests.get(url)
            response.encoding = response.apparent_encoding
            lines = response.text
            tmp = lines.split("患者報告数")
            num = tmp[1][:20].split("）")[0]
            ctext += num
            time.sleep(5) ###サイトに負荷をかけないように5秒間隔をあける
        except:
            ctext += ""
            num = ""
        print(ctext)    
        sumV += [ [ s_year, tweek, num] ]
df = pd.DataFrame(sumV)
print(df)
```
