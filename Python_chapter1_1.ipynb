### 変数の定義と簡単な演算

変数に値(実数値,整数)を代入してみます。  

プログラムでは通常等号(=)は代入に使用されます。   
値の代入は、等号(＝)の左に変数(値を入れたいものの名前)、右に値を書くことで行います。
[変数] = [値]

```
a = 2.0 
b = 5
```

これでa,bに値が代入されました。 次にこれらを使って、四則演算をしてみましょう


足し算

```
a+b
```

引き算

```
a-b
```

掛け算 掛け算記号は*(アスタリスク)です

```
a*b
```

演算した結果を別の変数cに代入して保存しておくこともできます。

```
c=a*b
```

このように、新たに値を代入した場合はc(=a*b)の値は表示されません。  
cの値が後で知りたいと思ったときは、以下のようにprint関数というものを使います。

```
print(c)
```

以下ではprint関数を逐次使って、値や演算の結果を確認することにしましょう。
また、各行に#(半角シャープ記号)をつけると、コンピュータは#以降に書かれたことを無視しますので、コメントを書くことができます。

なれないうちはどういった計算をするのか、その行をいつ編集したのかなど書いておくと便利かもしれません。

```
print(b//a) # 切り捨て除算
print(b%a) # bをaで割った余り
print(a**b) # aのb乗
```

```
あ=1.0;print(あ)
```

Pythonには、あらかじめ役割が与えられている言葉=予約語があり、以下は変数名として使う事はできません。

False, None, True, and, as, assert, break, class, continue, def, del, elif, else, except, finally, for, from, global, if, import, in, is, lambda, nonlocal,not, or, pass, raise, return, try, while, with, yield



これらは登場した際にその[役割]について説明することとしましょう。

真偽値 (bool,ブール)は　真(True)であるか偽(False)であるかの二通りの値しか取らない型です。

```
a=2.0; b = 5.0
```
